Construct color - absolute magnitude diagrams for the following star clusters: The Hyades (a young, nearby open cluster), M67 (an old, more distant open cluster), and NGC 6397 (a globlular cluster, at an even larger distance). Data can be found in gaiadr3.gaia_source catalog.

### Astrometric

Median uncertainty for bright source $G \lt 14$ mag is 0.03 mas (milli-arcseconds). Systematics (systematic uncertainties?) lower than 0.1 mas, parallax zeropoint error is about 0.03 mas.

### Photometric

Precision at $G = 12$ around 1 mmag (milli-mag) in three passbands, with systematics at level of 10 mmmag.

### Visibility periods

Set $visibility\_periods\_used > 8$ in query for reliable five-parameter solutions (Lindegren et al. 2018) and to remove outliers at the faint end of the local HRD (Arenou et al. 2018).

### Astrometic excess noise

The following filter from Lindegren et al 2018

$$ \sqrt{\frac{\chi^2}{\nu^{'} - 5}} \lt 1.2 \, max\left(1, \exp\left(-0.2\left(G - 19.5\right)\right)\right) $$

with $\chi^2$ and $\nu^{'}$ given as $astrometric\_chi2\_al$ and $astrometric\_n\_good\_obs\_al$, respectively, remove most artefacts between MS and WD, and observe genuine binaries on the HRDs.

### Absolute **Gaia** magnitude in **G** band

$$M_G = G + 5 + 5 \log_{10}\left(\omega/1000\right)$$

with parallax $\omega$ in milliarcseconds (mas). Apply a $10\%$ relative precision on parallax or an uncertainty on $M_G$ smaller than 0.22 mag: $parallax\_over\_error > 10$.

### Variable stars

Filters on relative flux error on $G$, $G_{BP}$, and $G_{RP}$ photometry to remove variable stars

$$phot\_g\_mean\_flux\_over\_error > 50$$

or $\sigma_G < 0.022$ mag

$$phot\_rp\_mean\_flux\_over\_error > 20$$
$$phot\_bp\_mean\_flux\_over\_error > 20$$

or $\sigma_{XP} < 0.054$ mag

### Treating blends in windows of blue and red photometers (BP and RP)

Sources within empirically defined locus of
$$\left(I_{BP} + I_{RP}\right)/I_G$$
fluxes ratio as a function of $G_{BP} - G_{RP}$.

$$phot\_bp\_rp\_excess\_factor < 1.3 + 0.06 \, \left(G_{BP} - G_{RP}\right)^2$$

### Hyades

From F. van Leeuwen 2009

Center: right ascension $\alpha_0 = 4^{\text{h}}26^{\text{m}}$ or $66.5^{\circ}$, declination $\delta_0 = 16^{\circ}54'$ or $16.9^{\circ}$ $\to \left(\alpha_0, \delta_0\right) = \left(66.5^{\circ},16.9^{\circ}\right)$. A search radius of 18.54$^{\circ}$/1112.4 arcmin, or 330 arcmin (wikipedia) can be applied based on the apparent dimensions of the cluster.

```python
# Hyades query without filters
# no proper motion cuts
hyades_q_nf = """SELECT DISTANCE(66.5, 16.9, ra, dec) AS ang_sep,
            phot_g_mean_mag AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(66.5, 16.9, ra, dec) < 1112.4/60.
            AND parallax < 21.53 + 2.76
            AND parallax > 21.53 - 2.76
"""

# Filtering out unreliable astrometry and photometry
# filters from Babusiaux et al. 2018, Section 2.1
hyades_q_f = """SELECT DISTANCE(66.5, 16.9, ra, dec) AS ang_sep,
            phot_g_mean_mag+5*log10(parallax)-10 AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(66.5, 16.9, ra, dec) < 1112.4/60.
            AND parallax_over_error > 10
            AND parallax < 21.53 + 2.76
            AND parallax > 21.53 - 2.76
            AND phot_g_mean_flux_over_error>50
            AND phot_rp_mean_flux_over_error>20
            AND phot_bp_mean_flux_over_error>20
            AND phot_bp_rp_excess_factor < 1.3+0.06*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND phot_bp_rp_excess_factor > 1.0+0.015*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND visibility_periods_used>8
            AND astrometric_chi2_al/(astrometric_n_good_obs_al-5)<1.44*greatest(1,exp(-0.4*(phot_g_mean_mag-19.5)))
"""
```

```python
cachedir = './joblib_cache'
# 1e7 nytes (10 MB)
# verbose=0: no verbose output during caching
memory = Memory(cachedir, verbose=0, bytes_limit=1e7)

@memory.cache
def get_gaia_query(q):
    start = time.time()
    job = Gaia.launch_job_async(q)
    print(f"Total time: {time.time()-start:0.2f} sec")
    return job.get_results()
```

```python
# Hyades data table
hyades_tbl_nf = get_gaia_query(hyades_q_nf)
hyades_tbl_f = get_gaia_query(hyades_q_f)

print(hyades_tbl_nf)
print(hyades_tbl_f)
```

## Fixed Age

```python
iso_path = "MIST_v1.2_vvcrit0.0_UBVRIplus/"

def process_data_file(directory, age):
    # List .iso.cmd files in directory
    files = [f for f in os.listdir(directory) if f.endswith(".iso.cmd")]
    
    processed_data_list=[]
    
    for filename in files:
        filepath = os.path.join(directory, filename)
        # read each file
        table = Table.read(filepath, format='ascii', header_start=-1)
        processed_data = process_data(table, age)
        processed_data_list.append(processed_data)
    
    return processed_data_list        
            
def process_data(table, age):
    processed_data = table[
        (table["log10_isochrone_age_yr"]==age)&
        (table["phase"]>=-1.0)&
        (table["phase"]<3.0)
    ]
    return processed_data

t_list = process_data_file(iso_path, 9.5)
```

```python
# Query without astrometry and photometry filters
# no proper motion cuts
NGC6397_q_nf = """SELECT DISTANCE(265.175385, -53.674335, ra, dec) AS ang_sep,
            phot_g_mean_mag AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(265.175385, -53.674335, ra, dec) < 32./60.
            AND parallax < 0.416 + 1.0
            AND parallax > 0.416 - 1.0
            AND pmra < 3.30 + 1.0
            AND pmra > 3.30 - 1.0
            AND pmdec < -17.60 + 1.0
            AND pmdec > -17.60 - 1.0
"""

# Filtering out unreliable astrometry and photometry
NGC6397_q_f = """SELECT DISTANCE(265.175385, -53.674335, ra, dec) AS ang_sep,
            phot_g_mean_mag+5*log10(parallax)-10 AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(265.175385, -53.674335, ra, dec) < 32./60.
            AND parallax_over_error > 10
            AND parallax < 0.416 + 1.0
            AND parallax > 0.416 - 1.0
            AND pmra < 3.30 + 1.0
            AND pmra > 3.30 - 1.0
            AND pmdec < -17.60 + 1.0
            AND pmdec > -17.60 - 1.0
            AND phot_g_mean_flux_over_error>50
            AND phot_rp_mean_flux_over_error>20
            AND phot_bp_mean_flux_over_error>20
            AND phot_bp_rp_excess_factor < 1.3+0.06*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND phot_bp_rp_excess_factor > 1.0+0.015*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND visibility_periods_used>8
            AND astrometric_chi2_al/(astrometric_n_good_obs_al-5)<1.44*greatest(1,exp(-0.4*(phot_g_mean_mag-19.5)))
"""
```

```python
NGC6397_tbl_nf = get_gaia_query(NGC6397_q_nf)
NGC6397_tbl_f = get_gaia_query(NGC6397_q_f)

print(NGC6397_tbl_nf)
print(NGC6397_tbl_f)
```
