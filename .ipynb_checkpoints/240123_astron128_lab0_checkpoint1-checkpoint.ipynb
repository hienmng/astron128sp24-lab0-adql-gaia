```python
import time
import os
import io
import requests
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import astropy.units as u
import matplotlib.pyplot as plt
from scipy.stats import tstd
from scipy.stats import norm
from scipy.stats import chi2
from scipy.stats import gaussian_kde
from scipy.optimize import curve_fit
from IPython.display import display, HTML
from astroquery.gaia import Gaia
from astropy.coordinates import SkyCoord
from joblib import Memory
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'

matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['legend.fontsize'] = 12
matplotlib.rcParams['figure.figsize'] = 8, 5
sns.set_style("whitegrid")
```

```python
# Hyades query without filters
# Hyades cluster's center and mean parallax from Leeuwen 2009
# Cone search radius from apparent dimension of 330 arcmin from wikipeda (need to obtain source)
# no proper motion cuts
hyades_q_nf = """SELECT DISTANCE(66.5, 16.9, ra, dec) AS ang_sep,
            phot_g_mean_mag AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(66.5, 16.9, ra, dec) < 330./60.
            AND parallax < 21.53 + 2.76
            AND parallax > 21.53 - 2.76
"""

# Filtering out unreliable astrometry and photometry
# filters from Babusiaux et al. 2018, Section 2.1
hyades_q_f = """SELECT DISTANCE(66.5, 16.9, ra, dec) AS ang_sep,
            phot_g_mean_mag+5*log10(parallax)-10 AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(66.5, 16.9, ra, dec) < 330./60.
            AND parallax_over_error > 10
            AND parallax < 21.53 + 2.76
            AND parallax > 21.53 - 2.76
            AND phot_g_mean_flux_over_error>50
            AND phot_rp_mean_flux_over_error>20
            AND phot_bp_mean_flux_over_error>20
            AND phot_bp_rp_excess_factor < 1.3+0.06*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND phot_bp_rp_excess_factor > 1.0+0.015*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND visibility_periods_used>8
            AND astrometric_chi2_al/(astrometric_n_good_obs_al-5)<1.44*greatest(1,exp(-0.4*(phot_g_mean_mag-19.5)))
"""
```

```python
cachedir = './joblib_cache'
# 1e7 nytes (10 MB)
# verbose=0: no verbose output during caching
memory = Memory(cachedir, verbose=0, bytes_limit=1e7)

@memory.cache
def get_gaia_query(q):
    start = time.time()
    job = Gaia.launch_job(q)
    print(f"Total time: {time.time()-start:0.2f} sec")
    return job.get_results()

# Hyades data table
hyades_tbl_nf = get_gaia_query(hyades_q_nf)
hyades_tbl_f = get_gaia_query(hyades_q_f)
```

```python
print(hyades_tbl_nf)
print(hyades_tbl_f)
```

# Color-Absolute-Magnitude

```python
"""
No astrometry/photometry quality filters
"""
hyades_absmag_nf = hyades_tbl_nf["mg"]
hyades_color_nf = hyades_tbl_nf["bp_rp"]

# Boolean mask for NaN values
hyades_nan_mask_nf = np.isnan(hyades_absmag_nf) | np.isnan(hyades_color_nf)

# Apply NaN mask to arrays
hyades_absmag_no_nan_nf = hyades_absmag_nf[~hyades_nan_mask_nf]
hyades_color_no_nan_nf = hyades_color_nf[~hyades_nan_mask_nf]

# Stack non-Nan arrays
hyades_values_nf = np.vstack([hyades_color_no_nan_nf,hyades_absmag_no_nan_nf])
# Calculate density with gaussian_kde
hyades_density_nf = gaussian_kde(hyades_values_nf)(hyades_values_nf)

# Plot highest density last
hyades_idx_nf = hyades_density_nf.argsort()
hyades_color_no_nan_nf = hyades_color_no_nan_nf[hyades_idx_nf]
hyades_absmag_no_nan_nf = hyades_absmag_no_nan_nf[hyades_idx_nf]
hyades_density_nf = hyades_density_nf[hyades_idx_nf]

"""
With astrometry/photometry quality filters
"""

hyades_absmag_f = hyades_tbl_f["mg"]
hyades_color_f = hyades_tbl_f["bp_rp"]

hyades_nan_mask_f = np.isnan(hyades_absmag_f) | np.isnan(hyades_color_f)

hyades_absmag_no_nan_f = hyades_absmag_f[~hyades_nan_mask_f]
hyades_color_no_nan_f = hyades_color_f[~hyades_nan_mask_f]

hyades_values_f = np.vstack([hyades_color_no_nan_f,hyades_absmag_no_nan_f])
hyades_density_f = gaussian_kde(hyades_values_f)(hyades_values_f)

hyades_idx_f = hyades_density_f.argsort()
hyades_color_no_nan_f = hyades_color_no_nan_f[hyades_idx_f]
hyades_absmag_no_nan_f = hyades_absmag_no_nan_f[hyades_idx_f]
hyades_density_f = hyades_density_f[hyades_idx_f]

fig0, ax0 = plt.subplots(1,2, figsize=(15,5))
scatter0 = ax0[0].scatter(hyades_color_no_nan_nf, hyades_absmag_no_nan_nf,  marker=".",c=hyades_density_nf)
ax0[0].invert_yaxis()
ax0[0].set_xlabel(r"G$_{BP}$ - G$_{BP}$")
ax0[0].set_ylabel(r"M$_{G}$")
ax0[0].set_title("Hyades Cluster w/o Astrometry/Photometry Quality Filters")
ax0[0].grid()
colorbar0 = plt.colorbar(scatter0, ax=ax0[0], label='Density')
scatter1 = ax0[1].scatter(hyades_color_no_nan_f, hyades_absmag_no_nan_f,  marker=".",c=hyades_density_f)
ax0[1].invert_yaxis()
ax0[1].set_xlabel(r"G$_{BP}$ - G$_{BP}$")
ax0[1].set_ylabel(r"M$_{G}$")
ax0[1].set_title("Hyades Cluster w/ Astrometry/Photometry Quality Filters")
ax0[1].grid()
colorbar1 = plt.colorbar(scatter1, ax=ax0[1], label='Density')
plt.show()
```

# Proper Motion Space

```python
hyades_pmra_nf = hyades_tbl_nf["pmra"]
hyades_pmdec_nf = hyades_tbl_nf["pmdec"]

hyades_pm_nan_mask_nf = np.isnan(hyades_pmra_nf) | np.isnan(hyades_pmdec_nf)

hyades_pmra_no_nan_nf = hyades_pmra_nf[~hyades_pm_nan_mask_nf]
hyades_pmdec_no_nan_nf = hyades_pmdec_nf[~hyades_pm_nan_mask_nf]

hyades_pm_values_nf = np.vstack([hyades_pmra_no_nan_nf, hyades_pmdec_no_nan_nf])
hyades_pm_density_nf = gaussian_kde(hyades_pm_values_nf)(hyades_pm_values_nf)

hyades_pm_idx_nf = hyades_pm_density_nf.argsort()
hyades_pmra_no_nan_nf = hyades_pmra_no_nan_nf[hyades_pm_idx_nf]
hyades_pmdec_no_nan_nf = hyades_pmdec_no_nan_nf[hyades_pm_idx_nf]
hyades_pm_density_nf = hyades_pm_density_nf[hyades_pm_idx_nf]

hyades_pmra_f = hyades_tbl_f["pmra"]
hyades_pmdec_f = hyades_tbl_f["pmdec"]

hyades_pm_nan_mask_f = np.isnan(hyades_pmra_f) | np.isnan(hyades_pmdec_f)

hyades_pmra_no_nan_f = hyades_pmra_f[~hyades_pm_nan_mask_f]
hyades_pmdec_no_nan_f = hyades_pmdec_f[~hyades_pm_nan_mask_f]

hyades_pm_values_f = np.vstack([hyades_pmra_no_nan_f, hyades_pmdec_no_nan_f])
hyades_pm_density_f = gaussian_kde(hyades_pm_values_f)(hyades_pm_values_f)

hyades_pm_idx_f = hyades_pm_density_f.argsort()
hyades_pmra_no_nan_f = hyades_pmra_no_nan_f[hyades_pm_idx_f]
hyades_pmdec_no_nan_f = hyades_pmdec_no_nan_f[hyades_pm_idx_f]
hyades_pm_density_f = hyades_pm_density_f[hyades_pm_idx_f]

fig1, ax1 = plt.subplots(1,2, figsize=(15,5))
scatter2 = ax1[0].scatter(hyades_pmra_no_nan_nf, hyades_pmdec_no_nan_nf, marker=".", c=hyades_pm_density_nf)
ax1[0].set_xlabel(r"$\mu_{\alpha^*}$ [mas/yr]")
ax1[0].set_ylabel(r"$\mu_{\delta^*}$ [mas/yr]")
ax1[0].set_title("Hyades Cluster w/o Astrometry/Photometry Quality Filters")
ax1[0].grid()
colorbar2 = plt.colorbar(scatter2, ax=ax1[0],  label='Density')
scatter3 = ax1[1].scatter(hyades_pmra_no_nan_f, hyades_pmdec_no_nan_f, marker=".", c=hyades_pm_density_f)
ax1[1].set_xlabel(r"$\mu_{\alpha^*}$ [mas/yr]")
ax1[1].set_ylabel(r"$\mu_{\delta^*}$ [mas/yr]")
ax1[1].set_title("Hyades Cluster w/o Astrometry/Photometry Quality Filters")
ax1[1].grid()
colorbar3 = plt.colorbar(scatter3, ax=ax1[1],  label='Density')
plt.show()
```
