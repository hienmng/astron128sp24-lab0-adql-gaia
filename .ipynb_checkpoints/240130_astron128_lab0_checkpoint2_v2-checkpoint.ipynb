```python
cachedir = './joblib_cache'
# 1e7 nytes (10 MB)
# verbose=0: no verbose output during caching
memory = Memory(cachedir, verbose=0, bytes_limit=1e7)

@memory.cache
def get_gaia_query(q):
    start = time.time()
    job = Gaia.launch_job(q)
    print(f"Total time: {time.time()-start:0.2f} sec")
    return job.get_results()
```

```python
# Hyades query without filters
# Hyades cluster's center and mean parallax from Leeuwen 2009
# no proper motion cuts
hyades_q_nf = """SELECT DISTANCE(66.5, 16.9, ra, dec) AS ang_sep,
            phot_g_mean_mag AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(66.5, 16.9, ra, dec) < 1112.4/60.
            AND parallax < 21.53 + 2.76
            AND parallax > 21.53 - 2.76
"""

# Filtering out unreliable astrometry and photometry
# filters from Babusiaux et al. 2018, Section 2.1
hyades_q_f = """SELECT DISTANCE(66.5, 16.9, ra, dec) AS ang_sep,
            phot_g_mean_mag+5*log10(parallax)-10 AS mg,
            bp_rp,ra,dec,source_id,parallax,pm,pmra,pmdec
        FROM gaiadr3.gaia_source
        WHERE DISTANCE(66.5, 16.9, ra, dec) < 1112.4/60.
            AND parallax_over_error > 10
            AND parallax < 21.53 + 2.76
            AND parallax > 21.53 - 2.76
            AND phot_g_mean_flux_over_error>50
            AND phot_rp_mean_flux_over_error>20
            AND phot_bp_mean_flux_over_error>20
            AND phot_bp_rp_excess_factor < 1.3+0.06*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND phot_bp_rp_excess_factor > 1.0+0.015*power(phot_bp_mean_mag-phot_rp_mean_mag,2)
            AND visibility_periods_used>8
            AND astrometric_chi2_al/(astrometric_n_good_obs_al-5)<1.44*greatest(1,exp(-0.4*(phot_g_mean_mag-19.5)))
"""
```

```python
# Hyades data table
hyades_tbl_nf = get_gaia_query(hyades_q_nf)
hyades_tbl_f = get_gaia_query(hyades_q_f)

# print(hyades_tbl_nf)
# print(hyades_tbl_f)
```

# MIST Isochrones

I eyeballed the proper motion cuts from the proper-motion-space plots in the previous checkpoint.

```python
# Proper Motion Cuts

hyades_tbl_pmcut_f = hyades_tbl_f[
    (hyades_tbl_f["pmra"]<180) &
    (hyades_tbl_f["pmra"]>50) &
    (hyades_tbl_f["pmdec"]>-100) &
    (hyades_tbl_f["pmdec"]<20)
]

# Color-AbsMag
# includes proper motion and quality filter cuts

hyades_absmag_f = hyades_tbl_pmcut_f["mg"]
hyades_color_f = hyades_tbl_pmcut_f["bp_rp"]

# Masking NaN values
hyades_nan_mask_f = np.isnan(hyades_absmag_f) | np.isnan(hyades_color_f)

hyades_absmag_no_nan_f = hyades_absmag_f[~hyades_nan_mask_f]
hyades_color_no_nan_f = hyades_color_f[~hyades_nan_mask_f]

# Creating density color bar
hyades_values_f = np.vstack([hyades_color_no_nan_f,hyades_absmag_no_nan_f])
hyades_density_f = gaussian_kde(hyades_values_f)(hyades_values_f)

# Sorting data by density
hyades_idx_f = hyades_density_f.argsort()
hyades_color_no_nan_f = hyades_color_no_nan_f[hyades_idx_f]
hyades_absmag_no_nan_f = hyades_absmag_no_nan_f[hyades_idx_f]
hyades_density_f = hyades_density_f[hyades_idx_f]
```

Fixed metallicity and age values, I used $\log\left(\text{age}\right) = 8.90$ and $Z = 0.020$ from Babusiaux et al 2018.

## Fixed Metallicity

```python
MIST = Table.read("hyades_MIST/MIST_iso_65b700c4e3770.iso.cmd", format="ascii", header_start=-1)

log_age = np.array([8.5, 8.6, 8.7, 8.8, 8.9, 9.0])

fig, ax = plt.subplots()
scatter = ax.scatter(hyades_color_no_nan_f, hyades_absmag_no_nan_f, marker=".", c="black", s=10)

for i, age in enumerate(log_age):
    # filter for MS, RGB, and age
    MIST_filtered = MIST[(MIST["log10_isochrone_age_yr"]==age)&
                        (MIST["phase"]>=-1.0)&(MIST["phase"]<3.0)]
    MIST_G = MIST_filtered["Gaia_G_EDR3"]
    MIST_BP_RP = MIST_filtered["Gaia_BP_EDR3"] - MIST_filtered["Gaia_RP_EDR3"]
    ax.plot(MIST_BP_RP, MIST_G, label=f"$\log$(age)={age}", c=plt.cm.tab10(i), linewidth=1.0)
    
ax.invert_yaxis()
ax.grid()
ax.legend(frameon=False)
ax.set_xlabel("BP - RP")
ax.set_ylabel(r"M$_{G}$")
ax.set_title("Z = 0.02")
plt.show()
```

## Fixed Age

```python
iso_path = "MIST_v1.2_vvcrit0.0_UBVRIplus/"

def process_data_file(directory):
    # List .iso.cmd files in directory
    files = [f for f in os.listdir(directory) if f.endswith(".iso.cmd")]
    
    processed_data_list=[]
    
    for filename in files:
        filepath = os.path.join(directory, filename)
        # read each file
        table = Table.read(filepath, format='ascii', header_start=-1)
        processed_data = process_data(table)
        processed_data_list.append(processed_data)
    
    return processed_data_list        
            
def process_data(table):
    processed_data = table[
        (table["log10_isochrone_age_yr"]==8.9)&
        (table["phase"]>=-1.0)&
        (table["phase"]<3.0)
    ]
    return processed_data

t_list = process_data_file(iso_path)
```

```python
cmap = plt.get_cmap('tab20')

fig, ax = plt.subplots()
scatter = ax.scatter(hyades_color_no_nan_f, hyades_absmag_no_nan_f, marker=".", c="black")

t_list_sorted = sorted(t_list, key=lambda x: x["[Fe/H]_init"][0])
t_list_sorted = t_list_sorted[-5:]

for i, processed_data in enumerate(t_list_sorted):
    Z_init = processed_data["[Fe/H]_init"][0]
    G = processed_data["Gaia_G_EDR3"]
    BP_RP = processed_data["Gaia_BP_EDR3"] - processed_data["Gaia_RP_EDR3"]
    color = cmap(i % 20)
    ax.plot(BP_RP, G, label=f"Z={Z_init}", color=color)
ax.invert_yaxis()
ax.grid()
ax.legend(frameon=False)
ax.set_xlabel("BP - RP")
ax.set_ylabel(r"M$_{G}$")
ax.set_title(r"$\log_{10}$(age)=8.90")
plt.show()
```
